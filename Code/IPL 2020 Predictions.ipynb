```python
import pandas as pd
import numpy as np
# we import matplotlib for visualisation
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as ticker
import matplotlib.ticker as plticker
# we import sk learn for splitting the dataset 
from sklearn.model_selection import train_test_split
# we import rfc for classification
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
from sklearn.preprocessing import LabelEncoder
```

```python
# we are loading the respective dataset 
IPL = pd.read_csv('Dataset.csv')
results = pd.read_csv('Match19.csv')
```

```python
# To see the dataset 
IPL.head(13)
```

```python
#Assignig the teams with their respective details from the other dataset
IPL_teams = ['Mumbai Indians',
'Chennai Super Kings',
'Kolkata Knight Riders',
'Sunrisers Hyderabad',
'Royal Challengers Banglore',
'Kings XI Punjab',
'Rajasthan Royals','Pune Warriors',
'Delhi Capitals','Rising Pune Supergiants','Gujarat Lions','Kochi Tuskers Kerala'
]
df_teams_1 = results[results['team1'].isin(IPL_teams)]
df_teams_2 = results[results['team2'].isin(IPL_teams)]
df_teams = pd.concat((df_teams_1, df_teams_2))
df_teams.drop_duplicates()
df_teams.count()
```

```python
#To see the final dataset 
df_teams.head(869)
```

```python
# Dropping the unwanted variables from the dataset ( the one not selected in the REP)
df_teams_2010 = df_teams.drop(['id','season','toss_winner','toss_decision','date','win_by_runs','result','dl_applied','Losser','win_by_wickets','player_of_match','umpire1','umpire2','umpire3'
], axis=1)
df_teams_2010.head()
```

```python
# Assign the variables for the winning teams 
df_teams_2010 = df_teams_2010.reset_index(drop=True)
df_teams_2010.loc[df_teams_2010.winner == df_teams_2010.team1,'winning_team']=1
df_teams_2010.loc[df_teams_2010.winner == df_teams_2010.team2, 'winning_team']=2
df_teams_2010 = df_teams_2010.drop(['winning_team'], axis=1)

df_teams_2010.head()
```

```python
# Get dummy variables
final = pd.get_dummies(df_teams_2010, prefix=['team1', 'team2','city','venue'], columns=['team1', 'team2','city','venue'])

# Separate X and y sets
X = final.drop(['winner'], axis=1)
y = final["winner"]
```

```python
# Splitting the dataset 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
final.head()
```

```python
# Importing for cross validation 
from sklearn.model_selection import KFold 
kf = KFold(n_splits=10, random_state=None) 
for train, test in kf.split(X):
    print("%s %s" % (train, test))
```

```python
# Fitting KNN classifier
from sklearn.neighbors import KNeighborsClassifier
Kn = KNeighborsClassifier(n_neighbors=5)

Kn.fit(X_train, y_train) 

score_1 = Kn.score(X_train, y_train)
score_2 = Kn.score(X_test, y_test)

print("Training set accuracy: ", '%.3f'%(score_1))
print("Test set accuracy: ", '%.3f'%(score_2))
```

```python
# Fitting the SVM classifier 
from sklearn import svm
svm = svm.SVC(kernel='rbf',C=1,gamma=1)

svm.fit(X_train, y_train) 

score_11 = svm.score(X_train, y_train)
score_22 = svm.score(X_test, y_test)

print("Training set accuracy: ", '%.3f'%(score_11))
print("Test set accuracy: ", '%.3f'%(score_22))
```

```python
# Fitting the Logistic Regression Classifier
from sklearn.linear_model import LogisticRegression
Lr = LogisticRegression(random_state=42)

Lr.fit(X_train,y_train)

score_a=Lr.score(X_train,y_train)
score_b=Lr.score(X_test,y_test)

print("Training set accuracy: ", '%.3f'%(score_a))
print("Test set accuracy: ", '%.3f'%(score_b))
```

```python
# Fitting the RandomForest Classifier
from sklearn.tree import export_graphviz
from IPython import display
rf = RandomForestClassifier(n_estimators=100, min_samples_leaf = 1 ,n_jobs = -2,oob_score=True,
                              random_state=42)
rf.fit(X_train, y_train) 

score = rf.score(X_train, y_train)
score2 = rf.score(X_test, y_test)

print("Training set accuracy: ", '%.3f'%(score))
print("Test set accuracy: ", '%.3f'%(score2))
```

```python
# Fitting Navie Bayes classifier
from sklearn.naive_bayes import GaussianNB

Dc = GaussianNB()

Dc.fit(X_train, y_train) 

score_aa = Dc.score(X_train, y_train)
score_bb = Dc.score(X_test, y_test)

print("Training set accuracy: ", '%.3f'%(score_aa))
print("Test set accuracy: ", '%.3f'%(score_bb))
```

```python
# Assigning variables for XGBoost 
seed = 15
test_size = 0.30
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=seed)
```

```python
# Fittting XGBoost Classifier
import xgboost as xgb

#from sklearn import mean_squared_error
model=xgb.XGBClassifier(learning_rate=0.8,max_depth=5,n_estimators=100,subsample = 1, min_child_weight=0.1,
                                gamma=0, colsample_bytree=0.8, objective= 'multi:softprob')
                                                                
model.fit(X_train,y_train)

score3 = model.score(X_train,y_train)
score4 = model.score(X_test,y_test)

print("Training set accuracy: ", '%.3f'%(score3))
print("Test set accuracy: ", '%.3f'%(score4))
```

```python
# Printing the parameters of XGBoost 
print('Parameters currently in used:\n')
print(model.get_params())
```

```python
# Loading the Testing Dataset for prediction
ranking = pd.read_csv('IPL_Rankings.csv') 
fixtures = pd.read_csv('Fixtures.csv')
```

```python
# Assign a variable to store the predictive outcome
pred_set = []
```

```python
# Adding the Columns to the primary dataset using the map function for position
fixtures.insert(1, 'first_position', fixtures['team1'].map(ranking.set_index('Team')['WinP']))
fixtures.insert(2, 'second_position', fixtures['team2'].map(ranking.set_index('Team')['WinP']))
```

```python
# Visual represention of data
fixtures = fixtures.iloc[:60, :]
fixtures.tail()
```

```python
# Adding the Columns to the primary dataset using the map function for rank 
fixtures.insert(3, 'Team_position', fixtures['team1'].map(ranking.set_index('Team')['Rank']))
fixtures.insert(4, 'team_position', fixtures['team2'].map(ranking.set_index('Team')['Rank']))
```

```python
fixtures = fixtures.iloc[:60, :]
fixtures.tail()
```

```python
# Calculating the position for team1 and team2
for index, row in fixtures.iterrows():
    if row['first_position'] < row['second_position'] and row['Team_position'] < row['team_position']:
        pred_set.append({'team1': row['team1'], 'team2': row['team2'],'city':row['city'],'venue':row['venue']})
    else:
        pred_set.append({'team1': row['team2'], 'team2': row['team1'],'city':row['city'],'venue':row['venue']})
        
pred_set = pd.DataFrame(pred_set)
backup_pred_set = pred_set
pred_set.head()
```

```python
# Calculating the output for the winner using matrix method
missing_cols = set(final.columns) - set(pred_set.columns)
for c in missing_cols:
    pred_set[c] = 0
pred_set = pred_set[final.columns]


pred_set = pred_set.drop(['winner'], axis=1)
pred_set.head()
```

```python
# Predicting the outcome
predictions = model.predict(pred_set)
for i in range(fixtures.shape[0]):
    print(backup_pred_set.iloc[i, 1] + " and " + backup_pred_set.iloc[i, 0])
    if predictions[i] == 1:
        print("winner: " + backup_pred_set.iloc[i, 1])
    
    else:
        print("winner: " + backup_pred_set.iloc[i, 0])
    print("")
```
